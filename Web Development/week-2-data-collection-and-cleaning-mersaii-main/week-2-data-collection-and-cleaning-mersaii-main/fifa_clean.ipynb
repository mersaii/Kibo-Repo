```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
fifa_dataset = pd.read_csv('./fifa_21_raw_data.csv')
```

```python
fifa_dataset.describe()
```

```python
# remove duplicates
fifa_dataset.drop_duplicates(subset=['Name'], inplace=True)
```

```python
fifa_dataset.describe()
```

```python
# conv feet and inches to cm
def feet_and_inches_to_cm(height_str):
    if "'" in height_str or '"' in height_str:
        height_str = height_str.strip('"')
        feet, inches = map(int, height_str.split("'"))
        return round(feet * 30.48 + inches * 2.54, 2)
    else:
        return height_str.strip('cm')
```

```python
# conv height to cm
fifa_dataset['Height'] = fifa_dataset['Height'].apply(feet_and_inches_to_cm)
```

```python
# conv lbs to kg
def lbs_to_kg(weight_str):
    if "lbs" in weight_str:
        weight_str = float(weight_str.strip('lbs'))
        weight_kg = round(weight_str * 0.45359237, 2)
        return weight_kg
    else:
        return weight_str.strip('kg')
```

```python
fifa_dataset['Weight'] = fifa_dataset['Weight'].apply(lbs_to_kg)
```

```python
# conv height and weight to numeric
fifa_dataset['Height'] = pd.to_numeric(fifa_dataset['Height'])
fifa_dataset['Weight'] = pd.to_numeric(fifa_dataset['Weight'])
fifa_dataset[['Height','Weight']].dtypes
```

```python
#replace '\n' with empty string in all columns
for column in fifa_dataset.columns:
    if fifa_dataset[column].dtype == 'object':  # Check if the column contains text data
        fifa_dataset[column] = fifa_dataset[column].str.replace('\n', '')

#rmv leading and trailing whitespaces in all text columns:
fifa_dataset = fifa_dataset.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
fifa_dataset.isnull().sum()
```

```python
# split long name column to new columns first name and last name 
fifa_dataset[['first name', 'last name']] = fifa_dataset['LongName'].str.split(' ', n=1, expand=True)
```

```python
# remove ouliers from hits column
fifa_dataset['Hits'] = fifa_dataset['Hits'].apply(lambda x: np.nan if 'K' in str(x) else x)
fifa_dataset['Hits'].isnull().sum()
```

```python
# conv hits column to numeric 
fifa_dataset['Hits'] = fifa_dataset['Hits'].astype(float)
```

```python
# replace nan values with mean value of hits column
mean = fifa_dataset['Hits'].mean()
# print(mean)
fifa_dataset['Hits'].fillna(fifa_dataset['Hits'].mean(), inplace=True)
# fifa_dataset['Hits'].isnull().sum()
```

```python
fifa_dataset['Contract'].isnull().sum()
```

```python
# convert ouliers to NAN in contract column
fifa_dataset['Contract'] = fifa_dataset['Contract'].apply(lambda x: np.nan if 'On' in str(x) or 'Free' in str(x) else x)
fifa_dataset['Contract'].isnull().sum()
```

```python
# extract end_year value from contract column
fifa_dataset[['end_year']]= fifa_dataset['Contract'].str.extract(r'~ (\d{4})')
fifa_dataset['end_year'].isnull().sum()
```

```python
# Define the condition to identify NAN values in the "Loan date end" column
condition = fifa_dataset['Loan Date End'].isna()

# Fill NaN values with 'Dec 31, ' + "end_year" for the selected rows
fifa_dataset.loc[condition, 'Loan Date End'] = 'Dec 31, ' + fifa_dataset['end_year'].astype(str)
# fifa_dataset['Loan Date End']
```

```python
# another fill method that worked
# fifa_dataset['Loan Date End'].fillna('Dec 31, ' + fifa_dataset['end_year'], inplace=True)
# fifa_dataset[['Loan Date End', 'end_year']]
# fifa_dataset['Contract']
```

```python
# Function to extract year
def extract_year(date_str):
    return date_str.split(', ')[1]

# fill NaN rows in the "Contract" column
fifa_dataset['Contract'] = fifa_dataset.apply(lambda row: row['Contract'] if not pd.isna(row['Contract']) else f"{extract_year(row['Joined'])} ~ {extract_year(row['Loan Date End'])}", axis=1)
```

```python
# Function to convert Value, Wage, and Release Clause to numeric
def convert_value_to_numeric(value):
    value = value.strip('â‚¬')
    a = '$'
    if value.endswith('M'):
        return a + str(float(value.replace('M', '')) * 1e6)
    elif value.endswith('K'):
        return a + str(float(value.replace('K', '')) * 1e3)
    else:
        return a + str(float(value))
```

```python
def strip_stars_and_convert_to_numeric(column_with_stars):
    # Function to strip stars/icons and convert the column to numeric
    return column_with_stars.str.replace('â˜…', '').astype(float)
```

```python
fifa_dataset.head()
```

```python
# clean value, wage and release clause columns
fifa_dataset['Value'] = fifa_dataset['Value'].apply(convert_value_to_numeric)
fifa_dataset['Wage'] = fifa_dataset['Wage'].apply(convert_value_to_numeric)
fifa_dataset['Release Clause'] = fifa_dataset['Release Clause'].apply(convert_value_to_numeric)
```

```python
fifa_dataset.head()
```

```python
fifa_dataset['W/F'] = strip_stars_and_convert_to_numeric(fifa_dataset['W/F'])
fifa_dataset['SM'] = strip_stars_and_convert_to_numeric(fifa_dataset['SM'])
fifa_dataset['IR'] = strip_stars_and_convert_to_numeric(fifa_dataset['IR'])
# fifa_dataset.head()
```

```python
# delete end_year column
fifa_dataset.drop(['end_year'], axis=1, inplace=True)
```

```python
# generate output file
fifa_dataset.to_csv('fifa_output.csv', index=False)
```
